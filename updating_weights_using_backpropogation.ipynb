```python
def initialise_parameters(layer_dim):
    np.random.seed(3)
    parameters={}
    l=len(layer_dim)
    
    for i in range(1,l):
        parameters['w'+str(i)]=np.ones((layer_dim[i-1],layer_dim[i]))*0.1
        parameters['b'+str(i)]=np.zeros((layer_dim[i],1))
    return parameters
```

```python
initialise_parameters([2, 2, 1])
```

```python
def linear_forward(a_prev,w,b):
    z=np.dot(w.T,a_prev)+b
    return z
```
